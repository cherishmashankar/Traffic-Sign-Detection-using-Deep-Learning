```python
def sign_predict(image):
    image = cv.resize(image, (32, 32))
    image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    model = load_model('german_traffic_sign_model.h5')
    image = np.array(image, dtype=np.float32)
    image = image/255
    image = np.reshape(image, (1, 32, 32))
    x = image.astype(np.float32)
    prediction = model.predict(x)
    prediction_max = np.argmax(prediction)
    prediction_label = labels_dict[prediction_max]
    confidence = np.max(prediction)
    return prediction_label, confidence
```

```python
image = cv.imread('GTSRB_Final_Test_Images/GTSRB/Final_Test/Images/00008.ppm')
plt.imshow(image)   
plt.show()        
sign, confidence = sign_predict(image)
print('Traffic Sign: ', sign)
print('Confidence: ', confidence)
```
